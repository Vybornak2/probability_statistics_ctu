```r
# Loading necessary libraries
library(moments)
library(nortest)
```

```r
# Reading data
x <- data.matrix(read.table("davkovac.txt"))
```

```r
# Summary
summary(x)
skewness(x)
kurtosis(x)
```

```r
# Scatter Plot
plot(x)
```

```r
# Boxplot
boxplot(x, main="Hmotnost balicku davkovanych automatem", ylab="hmotnost v dag")
```

```r
# Histograms and Density Analysis
hist(x)
hx <- hist(x)
print(hx) # Printing histogram frequency table
```

```r
# Probability density histogram with custom color and bins
hx <- hist(x)
print(hx)
h<-hist(x, main="Balici automat na kavu", xlab="hmotnost balicku v dekagramech", col="yellow", freq=FALSE, breaks = 8)
xfit<-seq(min(x),max(x),length=40)
yfit<-dnorm(xfit,mean=mean(x),sd=sd(x))
lines(xfit, yfit, col="red", lwd=2)
lines(density(x))
```

```r
# Empirical Cumulative Distribution Function (ECDF)
# plot(ecdf(x), verticals = FALSE, do.points="FALSE")
ecdf(x)
plot(ecdf(x), verticals = FALSE, do.points="FALSE")
xfit<-seq(min(x),max(x),length=40)
zfit<-pnorm(xfit,mean=mean(x),sd=sd(x))
lines(xfit, zfit, col="red", lwd=2)
```

```r
# Q-Q Plot (To visually check for normality)
qqnorm(x, main='Normal')
qqline(x)
```

```r
# Hypothesis tests to determine if data follows a normal distribution
shapiro.test(x)
ad.test(x)
jarque.test(x[,1])
```
