```r
# Data Loading and Normality Check
x <- data.matrix(read.table("davkovac.txt"))
qqnorm(x, main='Normal')
qqline(x)
```

```r
# Shapiro-Wilk normality test
shapiro.test(x)
```

```r
# Two-sided t-test
# Testing if the mean is equal to 25.0
test_res <- t.test(x, mu=25.0)
# test_res <- t.test(x, mu=25.0, alternative = "two.sided")
print(test_res)

boxplot(x)
ci <- test_res$conf.int
abline(h=ci[1], col="red", lwd=3)
abline(h=ci[2], col="red", lwd=3)
```

```r
    # One-sided t-tests
    # Testing if the mean is less than 25
    test_res <- t.test(x, mu=25, alternative = "less")
    print(test_res)

    boxplot(x)
    ci <- test_res$conf.int
    print(ci)
    abline(h=ci[1], col="red", lwd=3)
    abline(h=ci[2], col="red", lwd=3)
```

```r
# Testing if the mean is greater than 25
test_res <- t.test(x, mu=25, alternative = "greater")
print(test_res)
    
boxplot(x)
ci <- test_res$conf.int
print(ci)
abline(h=ci[1], col="red", lwd=3)
abline(h=ci[2], col="red", lwd=3)
```

```r
# Load data for two-sample tests
x <- data.matrix(read.table("dodavatelX.txt"))
y <- data.matrix(read.table("dodavatelY.txt"))
```

```r
# Boxplot comparison
boxplot(c(x), c(y))
```

```r
# Normality test for X
shapiro.test(x)
```

```r
# Normality test for Y
shapiro.test(y)
```

```r
# F-test for equality of variances
var.test(x, y)
```

```r
# Two-sample t-test (equal variances)
t.test(x, y, var.equal = T)
```

```r
# Read data for paired samples
# Pair t-test
cviceni <- data.frame(read.table("pred_cvicenim.txt"), read.table("po_cviceni.txt"))
cviceni[1:10, ]
```

```r
# Visual comparison of paired data
plot(cviceni[,1], cviceni[,2], xlab="pred cvicenim", ylab="po cviceni", ylim= c(5,23))
xfit <- seq(min(cviceni[,1]), max(cviceni[,1]), length=40)
lines(xfit, xfit, col = "red")
boxplot(cviceni[,1], cviceni[,2], names = c("pred cvicenim","po cviceni"))
```

```r
# Calculate differences
rozdil <- cviceni[,1] - cviceni[,2]
boxplot(rozdil, xlab = "rozdíl doby reakce")
```

```r
# Data Loading and Normality Check
qqnorm(rozdil, main='Normal')
qqline(rozdil)
qqline(rozdil, col = "red")
```

```r
# One-sample t-test on differences
t.test(rozdil)
```

```r
# Paired t-test
t.test(cviceni[,1], cviceni[,2], paired = T)
```

```r
# Wilcoxon signed-rank test (one sample)
wilcox.test(x, mu = 25, paired = F, correct = T)
```

```r
# Wilcoxon rank-sum test (two independent samples)
wilcox.test(cviceni[,1], cviceni[,2], paired = F)
```

```r
# Wilcoxon signed-rank test (paired)
wilcox.test(cviceni[,1], cviceni[,2], paired = T)
```

```r
# Sign test
binom.test(sum(rozdil > 0), length(rozdil), p=0.5, alternative="two.sided")
```
