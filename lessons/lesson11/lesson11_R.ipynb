```r
# Load data and visualize
pvl <- data.frame(read.table("pevnost_vlaken.txt", header=T))
boxplot(pevnost~dodavatel, data=pvl)
```

```r
# One-way ANOVA
oneway <- aov(pevnost~dodavatel, data=pvl)
summary(oneway)
```

```r
# Tukey HSD post-hoc test
TukeyHSD(oneway, conf.level = .95)
```

```r
# Pairwise t-test with Bonferroni correction
pairwise.t.test(pvl$pevnost, pvl$dodavatel, p.adj = "bonf")
```
