```r
# loading data
spotreba <- data.matrix(read.table("spotreba.txt"))
```

```r
# setting significance levels
a <- numeric(10)
for (i in 1:10) a[i] = 1-i/100
print(a)
```

```r
# initializing results matrix
clm=matrix(0, nrow = 10, ncol = 2)

# calculating interval estimates
for(i in 1:10) {
  a[i] = 1-i/100
  t=t.test(spotreba, conf.level=a[i])
  clm[i,]=c(t$conf.int)
}
print(clm)
```

```r
# visualization
plot(a, clm[,1], ylim=c(13.4, 14.4), type="l", 
     main="Intervalove odhady pro stredni hodnotu",
     xlab="alfa", ylab="stredni spotreba", col="blue")
lines(a, clm[,2], col="blue")

# declared consumption line
d <- numeric(10)
for (i in 1:10) d[i] = 13.5
lines(a, d, col="red")

# average measured consumption line
m <- numeric(10)
for (i in 1:10) m[i]=mean(spotreba)
lines(a, m)
```
